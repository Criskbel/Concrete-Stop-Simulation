# CONCRETE STOP #

## INTRODUCTION ##

In Mexico, the architectural design of certain cities is quite unique. One notable example is Guanajuato, where the city is built in a dome-like structure, with its main access points located at the top. As a result, when a renovation is needed in any of the lower constructions, materials must be transported from the upper levels down to the desired location, effectively following a vertical descent path.

Based on this scenario, we came up with the idea of designing a **mass-spring system** to assist construction workers in transporting concrete bags down to specific building areas. The mechanism works by placing a bag on the system from an upper level and allowing the spring to activate. The motion generated is then dampened by a shock-absorbing element (a damper), allowing the bag to slow down and stop safely, so that a worker at the lower level can collect it.


## OBJECTIVE ##

**CONCRETE STOP** aims to help construction workers transport material bags or objects weighing approximately 50 kg by using a mass-spring mechanism operated by a single person. The goal is to ensure the **safe descent of heavy objects**. Although this system may take slightly longer than a pulley, it offers greater **safety and stability** for the operator.

- Ensure the operator knows how to correctly use the system


## PROBLEM ##

A residential construction project is set to take place in Guanajuato, and we were assigned the responsibility of transporting materials — specifically cement bags.

To carry out this task, we are using a **mass-spring-damper system**. However, in order to ensure that our approach is both **efficient and safe**, we need to analyze the time it takes for the oscillation of the spring to **dampen and stabilize**, considering that each load will weigh approximately **50 kg**.

The differential equation that models our system is as follows:




```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import sympy as sys
from sympy import integrate
```

```python
k = 10.0
m = 50.0
B = -5.0

    
def resortin(A,t):
    return [A[1],-k*A[0]/m + B*A[1]/m]
inicial = [0.7,0.5]
tiempo = np.arange(0,150,0.01)
resultado = odeint(resortin,inicial,tiempo)
```

```python
fig, ax2 = plt.subplots(ncols=1, figsize = (10,10))

xx, yy = resultado.T

plt.plot(tiempo,xx)
plt.plot(tiempo,yy)
plt.show()
```

## CONCLUSION ##

In the field of engineering, it is often said that if you can model the world using mathematics, you're on the right track. That is why we approached this problem from a more realistic perspective. We used **differential equations** to solve a mass-spring system operating in a damped environment.

The resulting graph provides valuable information, allowing us to determine the proper coefficients for the system to behave **optimally and realistically**. If the damping force is reduced, the spring takes longer to settle; similarly, increasing the mass also extends the time to stabilization.

From the plotted function, we can conclude how the spring-mass system moved before fully coming to rest. According to our simulation, the material took **approximately 2 minutes** to reach the lower levels, which is acceptable and efficient. This method requires **fewer workers** compared to manually transporting materials from the rooftops of Guanajuato.
